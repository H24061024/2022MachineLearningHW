```python
import random
def shuffle(df,label):
    temp=list(range(len(df)))
    random.shuffle(temp)
    df=df.iloc[temp,:]
    label=label[temp]
    return(df,label)
```

```python
def updating_rule(beta,row,label):
    judge=0
    count=0
    while judge==0 and count <10:
        score=0
        for i in range(0,len(beta)):
            score=score+beta[i]*row[i]
        if (score>=0 and label==1):
            judge=1
        if (score<0 and label==-1):
            judge=1
        if (score>=0 and label==-1):
            for i in range(0,len(beta)):
                if i == (len(beta)-1):
                    if row[i]>=0:
                        beta[i]=beta[i]-30
                    else:
                        beta[i]=beta[i]+30
                else:
                    if row[i]>=0:
                        beta[i]=beta[i]-1
                    else:
                        beta[i]=beta[i]+1
        if (score<0 and label==1):
            for i in range(0,len(beta)):
                if i == (len(beta)-1):
                    if row[i]>=0:
                        beta[i]=beta[i]+30
                    else:
                        beta[i]=beta[i]-30
                else:
                    if row[i]>=0:
                        beta[i]=beta[i]+1
                    else:
                        beta[i]=beta[i]-1
        count=count+1
    return beta
```

```python
beta=[0]*31
```

```python
for i in range(0,len(label1)):
    beta=updating_rule(beta,data1.iloc[i],label1[i])
```

```python
def predict_value(beta,row):
    score=0
    for i in range(0,len(beta)):
        score=score+beta[i]*row[i]
        if (score>=0):
            judge=1
        if (score<0):
            judge=-1
    return judge
```

```python
from numpy.linalg import inv
def least_squares(df,label):
    temp_df=np.matmul(df.transpose(1,0),df)
    temp_inv=inv(np.matrix(temp_df))
    output_df=np.matmul(temp_inv,df.transpose(1,0))
    output_df=np.matmul(output_df,label)
    return output_df
```

```python
def vote_perceptron(df,label):
    iter_i=1
    random.seed(iter_i*6111016)
    label_total=[]
    while iter_i<=25:
        shuf_df,shuf_label=shuffle(df,label)
        beta=[0]*31
        for i in range(0,len(shuf_label)):
            beta=updating_rule(beta,shuf_df.iloc[i],shuf_label[i])
        label_pred=[]
        for i in range(0,len(label)):
            label_pred.append(predict_value(beta,df.iloc[i]))
        label_total.append(label_pred)
        iter_i=iter_i+1
    label_total=pd.DataFrame(label_total)
    mean_label=label_total.mean()
    out_label=[]
    for i in range(0,len(mean_label)):
        if(mean_label[i]>=0):
            out_label.append(1)
        else:
            out_label.append(-1)
    return out_label
```

```python
vote_pred=vote_perceptron(data_ori,label_ori)
```

```python
confusion_matrix(vote_pred,list(label_ori))
```

```python
def average_perceptron(df,label):
    iter_i=1
    random.seed(iter_i*6111016)
    beta_total=[]
    while iter_i<=25:
        shuf_df,shuf_label=shuffle(df,label)
        beta=[0]*31
        for i in range(0,len(shuf_label)):
            beta=updating_rule(beta,shuf_df.iloc[i],shuf_label[i])
        beta_total.append(beta)
        iter_i=iter_i+1
    beta_total=pd.DataFrame(beta_total)
    mean_beta=beta_total.mean()
    out_label=[]
    for i in range(0,len(label)):
        out_label.append(predict_value(mean_beta,df.iloc[i]))
    return out_label
```

```python
def minimize_linear_classifier(beta,df,label):
    random.seed(999)
    shuf_df,shuf_label=shuffle(df,label)
    dis_ori=np.array(shuf_df).dot(np.array(beta))
    dis_ori=np.absolute(dis_ori)
    dis_ori_sum=sum(dis_ori)/((np.array(beta).dot(np.array(beta)))**0.5)
    for k in range(0,40):
        for i in range(0,len(shuf_df)):
            score=0
            temp_beta=beta
            for j in range(0,len(beta)):
                score=score+beta[j]*shuf_df.iloc[i][j]
            if (score>=0 and shuf_label[i]==-1):
                for g in range(0,len(beta)):
                    if g == (len(beta)-1):
                        if shuf_df.iloc[i][j]>=0:
                            temp_beta[g]=temp_beta[g]-30
                        else:
                            temp_beta[g]=temp_beta[g]+30
                    else:
                        if shuf_df.iloc[i][j]>=0:
                            temp_beta[g]=temp_beta[g]-1
                        else:
                            temp_beta[g]=temp_beta[g]+1
                    dis_temp=np.array(shuf_df).dot(np.array(temp_beta))
                    dis_temp=np.absolute(dis_temp)
                    dis_temp_sum=sum(dis_temp)/((np.array(temp_beta).dot(np.array(temp_beta)))**0.5)
                    if(dis_temp_sum<=dis_ori_sum):
                        beta=temp_beta
            if (score<0 and shuf_label[i]==1):
                for g in range(0,len(beta)):
                    if g == (len(beta)-1):
                        if shuf_df.iloc[i][j]>=0:
                            temp_beta[g]=temp_beta[g]+30
                        else:
                            temp_beta[g]=temp_beta[g]-30
                    else:
                        if shuf_df.iloc[i][j]>=0:
                            temp_beta[g]=temp_beta[g]+1
                        else:
                            temp_beta[g]=temp_beta[g]-1
                    dis_temp=np.array(shuf_df).dot(np.array(temp_beta))
                    dis_temp=np.absolute(dis_temp)
                    dis_temp_sum=sum(dis_temp)/((np.array(temp_beta).dot(np.array(temp_beta)))**0.5)
                    if(dis_temp_sum<=dis_ori_sum):
                        beta=temp_beta
    return beta
```

```python
mini_beta = minimize_linear_classifier(ini_beta,data_ori,label_ori)
```

```python
def slack_minimize_linear_classifier(slack_C,beta,df,label):
    random.seed(9999999)
    shuf_df,shuf_label=shuffle(df,label)
    dis_ori=np.array(shuf_df).dot(np.array(beta))
    dis_ori=np.absolute(dis_ori)
    dis_ori_sum=0
    for i in range(0,len(dis_ori)):
        if (dis_ori[i])>slack_C:
            dis_ori_sum=dis_ori_sum+dis_ori[i]
    dis_ori_sum=dis_ori_sum/((np.array(beta).dot(np.array(beta)))**0.5)
    for k in range(0,40):
        for i in range(0,len(shuf_df)):
            score=0
            temp_beta=beta
            for j in range(0,len(beta)):
                score=score+beta[j]*shuf_df.iloc[i][j]
            if (score>=0 and shuf_label[i]==-1):
                for g in range(0,len(beta)):
                    if g == (len(beta)-1):
                        if shuf_df.iloc[i][j]>=0:
                            temp_beta[g]=temp_beta[g]-30
                        else:
                            temp_beta[g]=temp_beta[g]+30
                    else:
                        if shuf_df.iloc[i][j]>=0:
                            temp_beta[g]=temp_beta[g]-1
                        else:
                            temp_beta[g]=temp_beta[g]+1
                    dis_temp=np.array(shuf_df).dot(np.array(temp_beta))
                    dis_temp=np.absolute(dis_temp)
                    dis_temp_sum=sum(dis_temp)/((np.array(temp_beta).dot(np.array(temp_beta)))**0.5)
                    if(dis_temp_sum<=dis_ori_sum):
                        beta=temp_beta
            if (score<0 and shuf_label[i]==1):
                for g in range(0,len(beta)):
                    if g == (len(beta)-1):
                        if shuf_df.iloc[i][j]>=0:
                            temp_beta[g]=temp_beta[g]+30
                        else:
                            temp_beta[g]=temp_beta[g]-30
                    else:
                        if shuf_df.iloc[i][j]>=0:
                            temp_beta[g]=temp_beta[g]+1
                        else:
                            temp_beta[g]=temp_beta[g]-1
                    dis_temp=np.array(shuf_df).dot(np.array(temp_beta))
                    dis_temp=np.absolute(dis_temp)
                    dis_temp_sum=0
                    for h in range(0,len(dis_temp)):
                        if(dis_temp[h]>slack_C):
                            dis_temp_sum=dis_temp_sum+dis_temp[h]
                    dis_temp_sum=dis_temp_sum/((np.array(temp_beta).dot(np.array(temp_beta)))**0.5)
                    if(dis_temp_sum<=dis_ori_sum):
                        beta=temp_beta
    return beta
```
